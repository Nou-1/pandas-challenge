```python
#import dependencies
import pandas as pd
```

```python
#set path for file
filepath = "Resources/purchase_data.csv"

#read csv file
purchase_data = pd.read_csv(filepath)
```

```python
#display first 5 rows to obtain understanding of data
purchase_data.head()
```

Player Count

```python
#removing the duplicate data 
organised_players = purchase_data.loc[:,["Gender", "SN", "Age"]]
organised_players = organised_players.drop_duplicates()

#display the number of total players
total_players = organised_players.count()[0]

pd.DataFrame({"Total Players": [total_players]})
```

Purchasing Analysis (Total)

```python
#calculating number of unique items, average price of items, total purchases and total revenues
number_unique_items = purchase_data["Item ID"].nunique()
average_price = round(purchase_data["Price"].mean(), 2)
total_purchases = purchase_data["Item ID"].count()
total_revenue = purchase_data["Price"].sum()

#set information in a dataframe
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": number_unique_items,
               "Average Price": average_price,
               "Number of Purchases": total_purchases,
               "Total Revenue": [total_revenue]})

#applying $ sign and rounding to two decimal places
purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:,.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:.2f}".format)

#display dataframe
purchasing_analysis_df
```

Gender Demographics

```python
#calculating the total count and percentage of players per gender
gender_counts = organised_players["Gender"].value_counts()
gender_percentages = gender_counts/total_players*100

#set information in a dataframe
gender_demographics_df = pd.DataFrame({"Total Count": gender_counts,
                            "Percentage of Players": gender_percentages})

#applying decimal places and % formatting
gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].map("{:.2f}%".format)

#display dataframe
gender_demographics_df
```

Purchasing Analysis (Gender)

```python
#group data by Gender and Price column
gender_group = purchase_data.groupby(['Gender'])

#calculating purchase count, average price, total purchase, average total purchase per gender
group_purchase_count = gender_group["SN"].count()
group_average_price = gender_group["Price"].mean()
group_total_purchase = gender_group["Price"].sum()
group_average_total_purchase = group_total_purchase/gender_counts

#set information in dataframe
gender_purchasing_analysis_df = pd.DataFrame({"Purchase Count": group_purchase_count,
                                               "Average Purchase Price": group_average_price,
                                               "Total Purchase Price": group_total_purchase,
                                               "Average Total Purchase per Person": group_average_total_purchase})

#applying decimal places and $ sign formatting
gender_purchasing_analysis_df["Average Purchase Price"] = gender_purchasing_analysis_df["Average Purchase Price"].map("${:.2f}".format)
gender_purchasing_analysis_df["Total Purchase Price"] = gender_purchasing_analysis_df["Total Purchase Price"].map("${:.2f}".format)
gender_purchasing_analysis_df["Average Total Purchase per Person"] = gender_purchasing_analysis_df["Average Total Purchase per Person"].map("${:.2f}".format)

#display dataframe
gender_purchasing_analysis_df
```

Age Demographics

```python
#establish bins and corresponding labels
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99.90]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#organise data into bins
organised_age_ranges = pd.cut(organised_players["Age"], bins, labels=group_labels)
organised_total_count = organised_age_ranges.value_counts()
organised_percentage = organised_total_count/total_players * 100

#set information into a dataframe
Age_Demographics_df = pd.DataFrame({"Total Count": organised_total_count, "Percentage of Players": organised_percentage})

#sort data into index and apply formatting
Age_Demographics_df["Percentage of Players"] = Age_Demographics_df["Percentage of Players"].map("{:.2f}%".format)
Age_Demographics_df.sort_index()
```

Purchasing Analysis (Age)

```python
#establish bins and corresponding labels
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99.90]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#organise data into bins
age_ranges = pd.cut(purchase_data["Age"], bins, labels=group_labels)
age_purchase_count = purchase_data.groupby([age_ranges])["Price"].count()
age_average_purchase_price = purchase_data.groupby([age_ranges])["Price"].mean()
age_total_purchase_value = purchase_data.groupby([age_ranges])["Price"].sum()
age_average_total_purchase = age_total_purchase_value/organised_total_count

#set information into dataframe
purchasing_analysis_age_df = pd.DataFrame({"Purchase Count": age_purchase_count,
                                            "Average Purchase Price": age_average_purchase_price,
                                            "Total Purchase Price": age_total_purchase_value,
                                            "Average Total Purchase per Person": age_average_total_purchase})

#apply relevant formatting
purchasing_analysis_age_df["Average Purchase Price"] = purchasing_analysis_age_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_age_df["Total Purchase Price"] = purchasing_analysis_age_df["Total Purchase Price"].map("${:.2f}".format)
purchasing_analysis_age_df["Average Total Purchase per Person"] = purchasing_analysis_age_df["Average Total Purchase per Person"].map("${:.2f}".format)
purchasing_analysis_age_df.index.rename("Age Ranges", inplace=True)

#display dataframe
purchasing_analysis_age_df

```

Top Spenders

```python
#group by SN
group_players = purchase_data.groupby(['SN'])

#calculating purchase count, average purchase price, total purchase value
spenders_purchase_count = group_players["Item ID"].count()
spenders_average_purchase_price = round(group_players["Price"].mean(), 2)
spenders_total_purchase = group_players["Price"].sum()

#set information in a dataframe
top_spenders_analysis = pd.DataFrame({"Purchase Count": spenders_purchase_count,
                                      "Average Purchase Price": spenders_average_purchase_price,
                                      "Total Purchase Value": spenders_total_purchase})

#sorting results and applying formatting
top_5_spenders = top_spenders_analysis.sort_values(["Total Purchase Value"], ascending=False).head(5)
top_5_spenders["Average Purchase Price"] = top_5_spenders["Average Purchase Price"].map("${:.2f}".format)
top_5_spenders["Total Purchase Value"] = top_5_spenders["Total Purchase Value"].map("${:.2f}".format)

top_5_spenders
```

Most Popular Items

```python
#group by Item ID and Item Name
group_items = purchase_data.groupby(["Item ID","Item Name"])

#calculating purchase count, total purchase price and item price
items_purchase_count = group_items["Item ID"].count()
items_total_purchase_price = group_items["Price"].sum()
items_purchase_price = items_total_purchase_price/items_purchase_count

#set information in a dataframe
top_items_analysis = pd.DataFrame({"Purchase Count": items_purchase_count,
                                    "Item Price": items_purchase_price,
                                    "Total Purchase Value": items_total_purchase_price})

#sorting data and applying formatting
top_5_items = top_items_analysis.sort_values(["Purchase Count"], ascending=False).head(5)
top_5_items["Item Price"] = top_items_analysis["Item Price"].map("${:.2f}".format)
top_5_items["Total Purchase Value"] = top_items_analysis["Total Purchase Value"].map("${:.2f}".format)

top_5_items
```

Most Profitable Items

```python
#sort by total purchase value to obtain most profitable items
top_profitable_items = top_items_analysis.sort_values(["Total Purchase Value"], ascending=False).head(5)
top_profitable_items["Item Price"] = top_items_analysis["Item Price"].map("${:.2f}".format)
top_profitable_items["Total Purchase Value"] = top_items_analysis["Total Purchase Value"].map("${:.2f}".format)

top_profitable_items
```
